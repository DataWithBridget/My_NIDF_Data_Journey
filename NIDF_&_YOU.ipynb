#### **Amara**

```python
# Parameters
quarters = 22
quarterly_deposit = 100000
annual_rate = 0.09
quarterly_rate = annual_rate / 4  # 0.0225 per quarter

print(BOLD + "\n === Portfolio Simulation Begins === \n" + END)
print(f"Total Quarters: {quarters}")
print(f"Quarterly Deposit: ₦{quarterly_deposit}")
print(f"Quarterly Interest Rate: {quarterly_rate * 100:.2f}%\n")

# Scenario 1: Without Reinvestment (Simple Interest)
portfolio_no_reinvest = []
print(BOLD + "\n Calculating Portfolio WITHOUT Reinvestment (Simple Interest): \n" + END)
for k in range(quarters):
    principal = (k + 1) * quarterly_deposit
    interest = quarterly_deposit * quarterly_rate * ((k + 1) * (k + 2)) / 2
    total = principal + interest
    portfolio_no_reinvest.append(total)
    print(f"Quarter {k + 1}: Principal = ₦{principal}, Interest = ₦{interest:.2f}, Total = ₦{total:.2f}")
final_no_reinvest = portfolio_no_reinvest[-1]

# Scenario 2: With Reinvestment (Compounded Quarterly)
portfolio_reinvest = []
balance = 0
print(BOLD + "\n Calculating Portfolio WITH Reinvestment (Compound Interest): \n" + END)
for q in range(quarters):
    principal = (q + 1) * quarterly_deposit
    balance = (balance + quarterly_deposit) * (1 + quarterly_rate)
    portfolio_reinvest.append(balance)
    interest = balance - principal
    print(f"Quarter {q + 1}: Principal = ₦{principal}, Balance = ₦{balance}, Interest = ₦{interest:.2f}, Portfolio Balance = ₦{balance:.2f}")
final_reinvest = portfolio_reinvest[-1]

# Growth Metrics
total_saved = quarters * quarterly_deposit  # ₦2,200,000
growth_rate_no_reinvest = ((final_no_reinvest - total_saved) / total_saved) * 100
growth_rate_reinvest = ((final_reinvest - total_saved) / total_saved) * 100

print(BOLD + "\n === Summary === \n" + END)
print(BOLD + "Amara's Profile" + END)
print(f"Total Deposited: ₦{total_saved:,.2f}")
print(f"Portfolio Value WITHOUT Reinvestment: ₦{final_no_reinvest:,.2f}")
print(f"Total Interests Earned WITHOUT Reinvestment: ₦{(final_no_reinvest - total_saved):,.2f}")
print(f"Growth WITHOUT Reinvestment: {growth_rate_no_reinvest:.2f}%")
print(" ")
print(BOLD + "Alternate Amara's Profile" + END)
print(f"Portfolio Value WITH Reinvestment: ₦{final_reinvest:,.2f}")
print(f"Total Interests Earned WITHOUT Reinvestment: ₦{(final_reinvest - total_saved):,.2f}")
print(f"Growth WITH Reinvestment: {growth_rate_reinvest:.2f}%")
print(" ")
print(" ")
# Plot
quarters_list = list(range(1, quarters + 1))
plt.figure(figsize=(10, 6))
plt.plot(quarters_list, portfolio_no_reinvest, label='Without Reinvestment', marker='o')
plt.plot(quarters_list, portfolio_reinvest, label='With Reinvestment', marker='x')
plt.xlabel('Quarter Number')
plt.ylabel('Portfolio Value (₦)')
plt.title('Amara: Portfolio Growth Comparison (2020 Q1 - 2025 Q2)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### **The Aggressive Investor**

3. David

#### **David**


* Fund: 100% NIDF
* Reinvestment Strategy: Reinvest all dividend

```python
# nidf data
# Select only needed columns
nidf_data = nidf[['Year', 'Quarter', 'NAV per Unit (Ex Distribution)', 'Dividend (NGN)']]

# Reset index (optional)
nidf_data = nidf_data.reset_index(drop=True)

# Display the result
print(nidf_data)
```

```python
# Simulation function

def simulate_nidf_strategy(reinvest_dividends=True):
    total_units = 0
    investment_amount = 100_000
    pending_dividend = 0
    total_dividends_received = 0
    total_dividends_reinvested = 0

    portfolio_history = []

    for index, row in nidf_data.iterrows():
        quarter = f"{row['Year']} {row['Quarter']}"
        nav = row['NAV per Unit (Ex Distribution)']
        dividend_rate = row['Dividend (NGN)']

        # Calculate investment for this quarter
        invest_now = investment_amount
        if reinvest_dividends:
            invest_now += pending_dividend
            total_dividends_reinvested += pending_dividend
            pending_dividend = 0

        # Buy new units
        units_bought = invest_now / nav
        total_units += units_bought

        # Dividend earned on all units held at end of quarter
        dividend_earned = total_units * dividend_rate

        # Handle dividend
        if reinvest_dividends:
            if index == len(nidf_data) - 1:
                total_dividends_received += dividend_earned  # final quarter dividend is received
            else:
                pending_dividend = dividend_earned
        else:
            total_dividends_received += dividend_earned

        # Portfolio value at current NAV
        portfolio_value = total_units * nav

        # Track breakdown
        portfolio_history.append({
            "Quarter": quarter,
            "NAV": nav,
            "Investment": round(invest_now, 2),
            "Units Bought": round(units_bought, 4),
            "Total Units": round(total_units, 4),
            "Dividend Rate": dividend_rate,
            "Dividend Earned": round(dividend_earned, 2),
            "Cumulative Dividends Received": round(total_dividends_received, 2),
            "Portfolio Value": round(portfolio_value, 2)
        })

    # Summary
    total_invested = investment_amount * len(nidf_data)
    final_portfolio_value = total_units * nidf_data.iloc[-1]['NAV per Unit (Ex Distribution)'] # Access last row's NAV
    final_wealth = final_portfolio_value + total_dividends_received
    growth_rate = ((final_wealth - total_invested) / total_invested) * 100

    summary = {
        "Total Invested": round(total_invested, 2),
        "Final Portfolio Value": round(final_portfolio_value, 2),
        "Total Dividends Received": round(total_dividends_received, 2),
        "Total Dividends Reinvested": round(total_dividends_reinvested, 2),
        "Growth Rate (%)": round(growth_rate, 2)
    }

    return portfolio_history, summary

# Run both simulations
reinvesting_data, reinvesting_summary = simulate_nidf_strategy(reinvest_dividends=True)
no_reinvesting_data, no_reinvesting_summary = simulate_nidf_strategy(reinvest_dividends=False)

# Example output (summary)
print("With Reinvestment Summary:")
print(" ")
print(BOLD + "David's profile" + END)
for k, v in reinvesting_summary.items():
    print(f"{k}: ₦{v:,}")

print("\nWithout Reinvestment Summary:")
print(" ")
print(BOLD + "Alternate David's Profile" + END)
for k, v in no_reinvesting_summary.items():
    print(f"{k}: ₦{v:,}")
```
